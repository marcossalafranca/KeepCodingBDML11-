**Problema**: Implementar las operaciones de los multiconjuntos (utilizando las librer铆as y estructuras de datos vistas en el curso).

**Datos:**

Un multiconjunto es un conjunto en el que un elemento puede repetirse, es decir, cada elemento posee una multiplicidad (un n煤mero natural) que indica cu谩ntas veces el elemento es miembro del conjunto. Por ejemplo, en el multiconjunto `{a, a, b, b, b, c}`, las multiplicidades de los miembros a, b, y c son 2, 3, y 1, respectivamente.

Al igual que los conjuntos, poseen las siguientes caracter铆sticas y operaciones:
- Cardinalidad: indica el n煤mero de elementos del multiconjunto. Por ejemplo, la cardinalidad del multiconjunto `{a, a, b, b, b, c}` es 6 (la suma de sus multiplicidades).
- Inserci贸n: permite insertar una ocurrencia de un elemento en el multiconjunto.
- Eliminaci贸n: permite eliminar una ocurrencia de un elemento del multiconjunto.
- Comparaci贸n: compara dos multiconjuntos para determinar si son iguales.
- Pertenencia: determina si un elemento pertenece al multiconjunto.
- Subconjunto: determina si un multiconjunto es subconjunto de otro.
- Uni贸n: conjunci贸n de todos los elementos de dos multiconjuntos (sumando sus multiplicidades si un elementos est谩 en los dos).
- Intersecci贸n: elementos que est谩n en los dos multiconjuntos qued谩ndonos con la multiplicidad m谩s peque帽a.
- Diferencia: restar a un multiconjunto los elementos de otro.

```python
### TODO: Crear una funci贸n que dada una lista devuelva un multiconjunto
### El multiconjunto que devuelve puede crearse con la estructura de datos que se quiera (incluso una lista)
### TU RESPUESTA ABAJO

def multiconjunto(elementos):
    """
    Devuelve un multiconjunto con la lista de elementos dada

    Argumentos:
        elementos -- lista de elementos

    Ejemplo:
        elementos = [1,1,1,3,3,1,4,5,1,5]
    """
    # Crear diccionario vacio
    mc = {}
    # Recorrer los elementos y asignarles numero de repeticiones, si existe sumarle uno y si no existe darle valor 1
    for i in elementos:
        if i in mc:
            mc[i] += 1
        else:
            mc[i] = 1
    return mc

elementos = [1,1,1,3,3,1,4,5,1,5]
mc = multiconjunto(elementos)
mc

```

```python
### TODO: Crear una funci贸n que dado un multiconjunto devuelva su cardinalidad
### TU RESPUESTA ABAJO

def cardinalidad(multiconjunto):
    """
    Devuelve la cardinalidad del multiconjunto dado
    
    Argumentos:
        multiconjunto -- multiconjunto devuelto por la funci贸n creada anteriormente
    """
    # Sumar los valores
    return sum(multiconjunto.values())

print(cardinalidad(mc))

```

```python
### TODO: Crear una funci贸n que dado un multiconjunto y un elemento devuelva el multiconjunto con el elemento insertado
### TU RESPUESTA ABAJO

def inserta(multiconjunto, elemento):
    """
    Devuelve el multiconjunto habiendo insertado el elemento dado
    
    Argumentos:
        multiconjunto -- multiconjunto devuelto por la funci贸n creada anteriormente
        elemento -- elemento a insertar
    """
    # Comprobar que existe el elemento, si esta se le suma si no esta se le pone valor 1
    if elemento in multiconjunto:
        multiconjunto[elemento] += 1
    else:
        multiconjunto[elemento] = 1
    return multiconjunto

mc = inserta(mc, 9)
mc = inserta(mc, 9)
print(mc)
```

```python
### TODO: Crear una funci贸n que dado un multiconjunto y un elemento devuelva el multiconjunto con el elemento eliminado
### TU RESPUESTA ABAJO

def elimina(multiconjunto, elemento):
    """
    Devuelve el multiconjunto habiendo eliminado una ocurrencia del elemento dado
    
    Argumentos:
        multiconjunto -- multiconjunto devuelto por la funci贸n creada anteriormente
        elemento -- elemento a eliminar
    """
    # Comprobar que exista el elemento y restarle uno. Si su valor es cero eliminarlo con pop().
    if elemento in multiconjunto:
        multiconjunto[elemento] -= 1
        if multiconjunto[elemento] == 0:
            multiconjunto.pop(elemento, None)
    return multiconjunto

mc = elimina(mc, 9)
mc = elimina(mc, 9)

print(mc)
```

```python
### TODO: Crear una funci贸n que dados dos multiconjuntos devuelva su uni贸n
### TODO: Crear una funci贸n que dados dos multiconjuntos devuelva su intersecci贸n
### TODO: Crear una funci贸n que dados dos multiconjuntos devuelva su diferencia
### TU RESPUESTA ABAJO

def union(multiconjunto1, multiconjunto2):
    """
    Devuelve la uni贸n de los multiconjuntos
    
    Argumentos:
        multiconjunto1 -- multiconjunto devuelto por la funci贸n creada anteriormente
        multiconjunto2 -- multiconjunto devuelto por la funci贸n creada anteriormente
    """
    #Crear diccionario vacio
    union_mc = {}
    
    # Recorrer el primer conjunto y  a帽adirlo al nuevo diccionario
    for i, multi in multiconjunto1.items():
        union_mc[i] = multi
    # Recorrer el segundo conjunto
    for i, multi in multiconjunto2.items():
        # Si existe el elemento sumarle la repeticion sino  valor 1
        if i in union_mc:
            union_mc[i] += multi
        else:
            union_mc[i] = multi
    return union_mc

def interseccion(multiconjunto1, multiconjunto2):
    """
    Devuelve la intersecci贸n de los multiconjuntos
    
    Argumentos:
        multiconjunto1 -- multiconjunto devuelto por la funci贸n creada anteriormente
        multiconjunto2 -- multiconjunto devuelto por la funci贸n creada anteriormente
    """
    #Crear diccionario vacio
    interseccion_mc = {}
    
    # Recorrer el primer conjunto
    for i, multi in multiconjunto1.items():
        # Si se encuentra el elemento coger el elemento y su repetecion minima que seria 1
        if i in multiconjunto2:
            interseccion_mc[i] = min(multi, multiconjunto2[i])
    return multiconjunto(interseccion_mc)

def diferencia(multiconjunto1, multiconjunto2):
    """
    Devuelve la diferencia de los multiconjuntos
    
    Argumentos:
        multiconjunto1 -- multiconjunto devuelto por la funci贸n creada anteriormente
        multiconjunto2 -- multiconjunto devuelto por la funci贸n creada anteriormente
    """
    # Crear diccionario vacio para almacenar la diferencia
    diferencia_mc = {}

    # Recorrer los elementos del primer multiconjunto
    for i, multi in multiconjunto1.items():
        # Calcular la diferencia entre la multiplicidad del elemento en el primer y segundo multiconjunto
        diferencia = multi - multiconjunto2.get(i, 0)
        if diferencia > 0:
            # Agregar el elemento y su multiplicidad a la diferencia solo si su multiplicidad en multiconjunto1 es mayor que en multiconjunto2
            diferencia_mc[i] = diferencia
        elif i not in multiconjunto2:
            # Agregar el elemento y su multiplicidad a la diferencia si est谩 en multiconjunto1 pero no en multiconjunto2
            diferencia_mc[i] = multi

    return diferencia_mc

print (mc)
print(union(multiconjunto([1,3,4]), mc))
print(interseccion(multiconjunto([1,3,4]), mc))
print(interseccion(multiconjunto([1,3,4,3]), mc))
print(interseccion(multiconjunto([1,3,14,3]), mc))
print(diferencia(mc, multiconjunto([1,2,3,4])))
print(diferencia(mc, multiconjunto([1,3,3,3,4])))
print(diferencia(mc, multiconjunto([1,2,11,111])))
```

Deberas crear tu propia funci贸n para calcular el error de reconstrucci贸n, que viene definido por:

$$SSE =  \sum_{n}^{i=1}  \begin{Vmatrix}x_{i} -  \widehat{x}_i \end{Vmatrix} ^2 $$

Donde:

- $x_i$ son los valores de la matriz original X
- $\widehat{x}_i$ son los valores de la matriz reconstruida

```python
### TODO: Funci贸n para calcular el error de reconstrucci贸n
### TU RESPUESTA ABAJO

import numpy as np

def sse_score(X, X_hat):
    """
    Funci贸n para calcular el error de reconstrucci贸n
    
    Argumentos:
        X -- Matriz Original
        X_hat -- Matriz Reconstruida
        
    Ejemplo:
        X = np.array([[1, 2], [3, 4]])
        X_hat = np.array([[1.01, 1.75], [2.81, 3.99]])
        sse = sse_score(X, X_hat) # -> 0.09879
    """
    return np.sum((X - X_hat)**2)


X = np.array([[1, 2], [3, 4]])
X_hat = np.array([[1.01, 1.75], [2.81, 3.99]])
se = sse_score(X, X_hat) # -> 0.09879
se

```

Una vez que ya tenemos la funci贸n `sse` hecha, podemos pasar a construir la funci贸n que ejecutar谩 `SVM`.

```python
### TODO: Funci贸n para ejecutar SVM
### Tiene como entrada una matriz X
### Devuelve U, s, Vt

### Hint: S debe ser una matriz diagonal
### TU RESPUESTA ABAJO

def svm(X):
    """
    Funci贸n que ejecuta SVM y devuelve U, S, Vt

    Argumentos:
        X -- Matriz Original

    Ejemplo:
        X = np.array([[1, 2], [3, 4]])
        U, S, Vt = svm(X)

        # -> U = array([[-0.40455358, -0.9145143 ],
        #               [-0.9145143 ,  0.40455358]])
        # -> S = array([[5.4649857 , 0.        ],
        #               [0.        , 0.36596619]])
        # -> Vt = array([[-0.57604844, -0.81741556],
        #                [ 0.81741556, -0.57604844]])
    """
    U, s, Vt = np.linalg.svd(X, full_matrices=False)
    return U, np.diag(s), Vt

X = np.array([[1, 2], [3, 4]])
U, S, Vt = svm(X)

```

Como hemos visto en clase, las matrices obtenidas a partir de `SVM` nos sirven para reconstruir la matriz original `X`. Para ello, construye una funci贸n que permita reconstruir la matriz original `X` a partir de `U, s, Vt`.

```python
### TODO: Funci贸n para reconstruir la matriz original a partir de U, s, Vt
### Tiene como entrada U, s, Vt
### Devuelve X_hat
### TU RESPUESTA ABAJO


def reconstruction(U, S, Vt):
    """
    Funci贸n que reconstruye la matriz original a partir de U, s, Vt
    
    Argumentos:
        U -- Matriz de Singular Vectors
        s -- Matriz de Eigenvalues
        Vt -- Matriz de Singular Vectors
        
    Ejemplo:
        U = np.array([[-0.40455358, -0.9145143 ],
                      [-0.9145143 ,  0.40455358]])
        S = np.array([[5.4649857 , 0.        ],
                      [0.        , 0.36596619]])
        Vt = np.array([[-0.57604844, -0.81741556],
                       [ 0.81741556, -0.57604844]])
        X_hat = reconstruction(U, S, Vt)
        
        # X_hat -> array([[0.99999999, 1.99999998],
        #                 [3.00000003, 4.00000001]])
    """
    return U.dot(S.dot(Vt))

U = np.array([[-0.40455358, -0.9145143 ],
                      [-0.9145143 ,  0.40455358]])
S = np.array([[5.4649857 , 0.        ],
                      [0.        , 0.36596619]])
Vt = np.array([[-0.57604844, -0.81741556],
                       [ 0.81741556, -0.57604844]])
X_hat = reconstruction(U, S, Vt)
        
X_hat
```

Calcula el error de reconstrucci贸n usando la funci贸n `sse` que has programado anteriormente.

```python
sse = sse_score(X, X_hat)
print(f"El error de reconstrucci贸n es: {sse}")
```

Una vez que hemos programado todas las funciones necesarias para realizar `SVM` y medir el error de reconstrucci贸n, podemos proceder a realizar la compresi贸n de la imagen. Esta [p谩gina web](http://timbaumann.info/svd-image-compression-demo/) te ayudar谩 a repasar y a entender como calcular la compresi贸n.

```python
### TODO: Funci贸n que recibe una imagen A y devuelve la imagen comprimida
### Tiene como entrada A y el n煤mero de componentes para realizar la reducci贸n de dimensionalidad
### Devuelve la imagen comprimidad, el error de reconstrucci贸n y el ratio de compresi贸n

### Hint: Usa las funciones anteriormente construidas
### TU RESPUESTA ABAJO


def image_compression(A, n_components):
    """
    Funci贸n para comprimir una imagen A
    
    Argumentos:
        A -- Imagen original
        n_components -- N煤mero de componentes
        
    Ejemplo:
        A_hat, sse, comp_ratio = image_compression(A, n_components=50)
    """
    # Sacamos U, S y Vt de la imagen comprimida
    U_comp, S_comp, Vt_comp = svm(A)

    # Hacemos la recosntruccion de la matriz a partir de U, S y Vt
    A_hat_comp = reconstruction(U_comp, S_comp, Vt_comp)

    # Calculamos el error de reconstruccion
    sse_comp   = sse_score(A, A_hat_comp)

    # Ratio de compresi贸n mr + r + rn
    ratio_comp = (A.shape[1]*n_components + n_components + A.shape[0]*n_components) / (A.shape[1] * A.shape[0])

    return A_hat_comp, sse_comp, ratio_comp

A_hat, sse, comp_ratio = image_compression(A, n_components=50)

print("La imagen comprimida es: ", plt.imshow(A_hat, cmap=plt.cm.gray))
print("El error de reconstrucci贸n es: ", sse)
print("El ratio de compresi贸n es: ", comp_ratio)
```

```python
### TODO: Funci贸n para crear un nuevo subset en base a valores
### Como argumento recibe un dataframe y una lista de tuples
### Tuples: (column_name, min_value, max_value)
### Devuelve un dataframe que excluye las filas donde los valores, en la columna que le hemos indicado, exceden los valores
### que le hemos indicado
### No eliminar la fila si los valores son iguales al min/max valor
### TU RESPUESTA ABAJO

def column_cutoff(data_frame, cutoffs):
    """Crea un nuevo dataframe en base a unos l铆mites
    
    Argumentos:
        data_frame -- Dataframe Object
        cutoffs -- Lista de tuples con el siguiente formato:
        (column_name, min_value, max_value)
        
    Ejemplo:
        data_frame = read_into_data_frame('train.csv')
        # Remove data points with SalePrice < $50,000
        # Remove data points with GrLiveAre > 4,000 square feet
        cutoffs = [('SalePrice', 50000, 1e10), ('GrLivArea', 0, 4000)]
        selected_data = column_cutoff(data_frame, cutoffs)
    """
    # Seleccionar las columnas especificadas en los tuples
    data_frame = data_frame[[i[0] for i in cutoffs]]
    # Filtrar las filas donde los valores est谩n fuera del rango especificado para cada columna
    for i in cutoffs:
        # Filtrar las filas donde el valor de la columna es menor que el valor m铆nimo
        data_frame = data_frame[data_frame[i[0]] > i[1]]
        # Filtrar las filas donde el valor de la columna es mayor que el valor m谩ximo
        data_frame = data_frame[data_frame[i[0]] < i[2]]
    return data_frame

data_frame = read_to_df('./data/train.csv')
# Remove data points with SalePrice < $50,000
# Remove data points with GrLiveAre > 4,000 square feet
cutoffs = [('SalePrice', 50000, 1e10), ('GrLivArea', 0, 4000)]
selected_data = column_cutoff(data_frame, cutoffs)
print(selected_data)
```

#### M铆nimos Cuadrados / Least Squares

Ahora, implementar谩s la ecuaci贸n $w_{LS}$:

<center>$w_{LS} = (X^T X)^{1}X^T y,$</center>

```python
### TODO: Funci贸n para resolver la ecuaci贸n wLS
### Toma como argumentos dos matrices, una para X y otra para y
### Asumimos que las matrices tienen las dimensiones correctas

### Paso 1: Asegurate que n > d. 
### Es decir, que el n煤mero de observaciones es mayor que el n煤mero de dimensiones.
### O lo que es lo mismo, que el n煤mero de filas de cada matriz sea mayor que el n煤mero de columnas
### Si no es as铆, debes transponer las matrices

### Paso 2: Debes a帽adir a la matriz X un vector columna del tama帽o (n x 1)

### Paso 3: Usa la ecuaci贸n de arriba para obtener wLS

### TU RESPUESTA ABAJO


def least_squares_weights(input_x, target_y):
    """Resuelve la ecuaci贸n para wLS
    
    Argumentos:
        input_x -- Matriz con los datos de entrenamiento
        target_y -- Vector con los datos de salida
        
    Ejemplo:
        import numpy as np
        training_y = np.array([[208500, 181500, 223500, 
                                140000, 250000, 143000, 
                                307000, 200000, 129900, 
                                118000]])
        training_x = np.array([[1710, 1262, 1786, 
                                1717, 2198, 1362, 
                                1694, 2090, 1774, 
                                1077], 
                               [2003, 1976, 2001, 
                                1915, 2000, 1993, 
                                2004, 1973, 1931, 
                                1939]])
        weights = least_squares_weights(training_x, training_y)
        
        print(weights)  #--> np.array([[-2.29223802e+06],
                        #              [ 5.92536529e+01],
                        #              [ 1.20780450e+03]])
                           
        print(weights[1][0]) #--> 59.25365290008861
    
    Asumimos:
        -- target_y es un vector con el mismo n煤mero de observaciones que input_x
    """
    # Paso 1: Reorganizar input_x y target_y si es necesario
    n = 0
    if input_x.shape[0] < input_x.shape[1]:
        input_x = np.transpose(input_x)
        n = input_x.shape[0]
        
    if target_y.shape[0] < target_y.shape[1]:
        target_y = np.transpose(target_y)

    # Paso 2: Agregar una columna de unos a la matriz de entrada
    X = np.concatenate((np.ones((n, 1)), input_x), axis=1)

    # Paso 3: Calcular los pesos 贸ptimos utilizando la ecuaci贸n de m铆nimos cuadrados
    dot1   = np.dot(X.T, X)
    linal  = np.linalg.inv(dot1)
    dot2   = np.dot(X.T, target_y)
    dot3   = np.dot(linal, dot2)
    
    return dot3
    
import numpy as np
training_y = np.array([[208500, 181500, 223500, 
                                140000, 250000, 143000, 
                                307000, 200000, 129900, 
                                118000]])
training_x = np.array([[1710, 1262, 1786, 
                                1717, 2198, 1362, 
                                1694, 2090, 1774, 
                                1077], 
                               [2003, 1976, 2001, 
                                1915, 2000, 1993, 
                                2004, 1973, 1931, 
                                1939]])
weights = least_squares_weights(training_x, training_y)

print(weights)  #--> np.array([[-2.29223802e+06],
                        #              [ 5.92536529e+01],
                        #              [ 1.20780450e+03]])

print(weights[1][0]) #--> 59.25365290008861
```

## 4. Linear Regression - Gradient Descent

En este ejercicio resolveras el mismo problema anterior pero usando **Gradient Descent**

**Objetivos**:
- Asegurar los fundamentos matem谩ticos detr谩s del Gradient Descent.

**Problema**: Usando datos sobre el precio de la vivienda, intentaremos predecir el precio de una casa en base a la superficie habitable con un modelo de regresi贸n.

**Datos:** [Kaggle's House Prices Dataset](https://www.kaggle.com/c/house-prices-advanced-regression-techniques/data)

**Repaso:**

$$ RSS(w) = \sum_{n=1}^{N}[y_n-f(x_n)]^2 =  \sum_{n=1}^{N}[y_n- (w_0 + \sum_{d=1}^{D}w_dx_{nd}) ]^2 .$$

Loss function:

$$ RSS(w) = \frac{1}{2}\sum_{n=1}^{N}[y_n-f(x_n)]^2$$

Y lo que queremos es minimizar esta distancia, para que el modelo se acerque lo m谩ximo posible a los valores verdaderos.

$$\nabla RSS(w) = X^T(Xw^t-y)$$

En resumen, el gradient descendiente para una regresi贸n lineal, se basa en resolver esta ecuaci贸n de forma iterativa:

$$w^{t+1} = w^t - \eta * \nabla RSS(w)$$

#### Leer Datos

```python
import pandas as pd
import numpy as np

# Leer datos
data = pd.read_csv('./data/train.csv')

# Extraer dichas columnas
newData = data[['GrLivArea','SalePrice']]
print(newData.head())

# Contruir x - y
x = newData['GrLivArea']
y = newData['SalePrice']

# Standarizar los datos
x = (x - x.mean()) / x.std()
x = np.c_[np.ones(x.shape[0]), x] 

print("Shape of X: ", x.shape)
print("Shape of y:", y.shape)
```

#### Gradient Descent

```python
### TODO: Funci贸n para encontrar los valores w usando Gradient Descent
### Toma como argumentos: X, y, w, n_iterations, eta
### Completa la funci贸n a帽adiendo la loss funci贸n y la updating rule
### TU RESPUESTA ABAJO

def gradient_descent(x, y, w, iterations, eta):
    """Gradient descent
    
    Argumentos:
        x -- Matriz con los datos de entrenamiento
        y -- Vector con los datos de salida
        w -- Vector aleatoriamente inicializado
        iterations -- N煤mero de iteraciones
        eta -- Learning Rate
        
    Ejemplo:
        import numpy as np

        # Learning rate
        eta = 0.01 

        # N煤mero de iteraciones
        iterations = 2000 #No. of iterations

        # Seed para inicializar w
        np.random.seed(123)
        w0 = np.random.rand(2)
        
        training_y = np.array([208500, 181500, 223500, 
                                140000, 250000])
        training_x = np.array([[ 1.        ,  0.37020659],
                               [ 1.        , -0.48234664],
                               [ 1.        ,  0.51483616],
                               [ 1.        ,  0.38352774],
                               [ 1.        ,  1.29888065]])
                            
        weights, loss = gradient_descent(training_x, training_y, w0, iterations, eta)
        
        print(weights[-1])  #--> np.array([183845.82320222  40415.66453324])
    """
    # Inicializar las listas donde se almacenar谩n los valores pasados de  loss y los pesos w
    p_loss = []
    p_w = [w]


    # Iterar "iterations" veces
    for i in range(iterations):
        # Calcular la predicci贸n usando el producto punto entre x y w
        prediction = np.dot(x, w)
        # Calcular el error restando la predicci贸n y el vector y
        error = prediction - y

        # Calcular la funci贸n de p茅rdida y almacenarla en la lista de p_loss
        loss = np.dot(error.T, error) / 2
        p_loss.append(loss)

        GradRss = np.dot(x.T, error) * eta

         # Calcular el gradiente de la funci贸n de p茅rdida y actualizar los pesos w
        w -= GradRss
        p_w.append(w)

    return p_w, p_loss

import numpy as np

# Learning rate
eta = 0.01

# N煤mero de iteraciones
iterations = 2000 #No. of iterations

# Seed para inicializar w
np.random.seed(123)
w0 = np.random.rand(2)

training_y = np.array([208500, 181500, 223500,
                                140000, 250000])
training_x = np.array([[ 1.        ,  0.37020659],
                               [ 1.        , -0.48234664],
                               [ 1.        ,  0.51483616],
                               [ 1.        ,  0.38352774],
                               [ 1.        ,  1.29888065]])

weights, loss = gradient_descent(training_x, training_y, w0, iterations, eta)

print(weights[-1])  #--> np.array([183845.82320222  40415.66453324])
```

```python
import re
import lxml
from bs4 import BeautifulSoup
import urllib
import urllib.request

url = "https://es.wikipedia.org/wiki/Canis_lupus_familiaris"

def parse_from_url(url):
    """
    Funci贸n para extraer el contenido (raw text) de una p谩gina web
    """
    
    html = urllib.request.urlopen(url).read()
    soup = BeautifulSoup(html, "html.parser" )
    for script in soup(["script", "style"]):
        script.extract()
        
    text = soup.get_text()
    
    # Eliminar saltos de linea
    text = re.sub('\s+', ' ', text)
    return text

parse_from_url(url)[:1000]
```

```python
### TODO: Escribe una funci贸n que recibe una lista de N p谩ginas web
### Y calcula todas las combinaciones posibles entre ellas, no importa el orden
### [web_1, web_2, web_3, ...]
### Devuelve una lista de tuples con las combinaciones [(web_1, web_2), (web_1, web_3), ...]

# HINT: Puedes implementar esta funci贸n como quieras pero la librer铆a itertools 
#       proporciona una funci贸n llamada `combinations` para realizar esta tarea.

### TU RESPUESTA ABAJO
import itertools
    
def combinations(url_ls):
    """Calcula todas las combinaciones posibles entre los elementos de una lista
    
    Argumentos:
        url_ls -- Lista de urls
        
    Ejemplo:
        url_ls = ['https://es.wikipedia.org/wiki/Canis_lupus_familiaris', 
        'https://es.wikipedia.org/wiki/Canis_lupus',
        'https://es.wikipedia.org/wiki/Felis_silvestris_catus']
        
        permutation = combinations(url_ls)  
    
    Requerimientos:
        Puedes implementar esta funci贸n como quieras pero la librer铆a itertools 
        proporciona una funci贸n llamada `combinations` para realizar esta tarea.
    """
    return list(itertools.combinations(url_ls, 2))

url_ls = ['https://es.wikipedia.org/wiki/Canis_lupus_familiaris', 
        'https://es.wikipedia.org/wiki/Canis_lupus',
        'https://es.wikipedia.org/wiki/Felis_silvestris_catus']

combinations(url_ls)
```

```python
### TODO: Escribe una funci贸n que recibe una lista con tuples
### [({'que', 'hola'}, {'que', 'es', 'guay'}), ({'que', 'hola'}, {'madrid', 'la', 'es'})]
### Y devuelve una lista con la union de los conjuntos
### [({'que', 'hola', 'es', 'guay'}), ({'que', 'hola', 'madrid', 'la', 'es'})]
### TU RESPUESTA ABAJO

def union(comb_ls):
    """Calcula la uni贸n por cada tuple de una lista 

    Argumentos:
        comb_ls -- Lista de tuples

    Ejemplo:
        comb_ls = [({'que', 'hola'}, {'que', 'es', 'guay'}), ({'que', 'hola'}, {'madrid', 'la', 'es'})]

        union_ls = union(comb_ls)  # -> [{'es', 'que', 'hola', 'guay'}, {'es', 'que', 'hola', 'madrid', 'la'}]
    """
    # Crear lista
    lista_union = []
    # Recorrer  lista
    for i in comb_ls:
        # Unir el priner set del tuple con el segundo set
        union_set = i[0] | i[1]
        # Agregar  el resultado a union
        lista_union.append(union_set)
    return lista_union

comb_ls = [({'que', 'hola'}, {'que', 'es', 'guay'}), ({'que', 'hola'}, {'madrid', 'la', 'es'})]

union_ls = union(comb_ls)  # -> [{'es', 'que', 'hola', 'guay'}, {'es', 'que', 'hola', 'madrid', 'la'}]
union_ls
```

Una vez que tenemos una lista de conjuntos por cada par de p谩ginas web, podemos convertir el texto de la p谩gina web a un vector.

```python
def set2vector(tokens_web1, tokens_web_2, set_web1, set_web2):
    """
    Funci贸n para convertir un conjunto a vector
    
    Argumentos:
        tokens_web1 -- Contenido tokenizado de p谩gina web 1
        tokens_web_2 -- Contenido tokenizado de p谩gina web 2
        set_web1 -- Conjunto de palabras de la p谩gina web 1
        set_web2 -- Conjunto de palabras de la p谩gina web 2
        
    Ejemplo:
        tokens_web1 = ["hola", "que", "tal", "soy", "Alex"]
        tokens_web_2 = ["hola", "me", "llamo"]
        set_web1 = {"hola", "que", "tal"}
        set_web2 = {"hola", "me", "llamo"}
        union_ls = set2vector(tokens_web1, tokens_web_2, set_web1, set_web2)  
        
    Requerimientos:
        Depende de la funci贸n `union()`
    """
    
    # Unimos los conjuntos
    join_set = union([(set_web1, set_web2)])[0]
    
    # Creamos dos listas
    web1_array = []
    web2_array = []

    # Recorrer la union  y comprobar si estan en los token_web
    for i in join_set:

        if i in tokens_web1:
            web1_array.append(1)
        else:
            web1_array.append(0)

        if i in tokens_web_2:
            web2_array.append(1)
        else:
            web2_array.append(0)

    return web1_array, web2_array

tokens_web1 = ["hola", "que", "tal", "soy", "Alex"]
tokens_web_2 = ["hola", "me", "llamo"]
set_web1 = {"hola", "que", "tal", "soy", "Alex"}
set_web2 = {"hola", "me", "llamo"}
web1_array, web2_array = set2vector(tokens_web1, tokens_web_2, set_web1, set_web2) 

```

### Cosine Similarity

Por 煤ltimo, ya podemos implementar la ecuaci贸n: $\cos \theta = \frac{\overrightarrow{u} \cdot \overrightarrow{v}}{|\overrightarrow{u}||\overrightarrow{v}|}$

```python
### TODO: Escribe una funci贸n que recibe dos vectores, u y v
### Y devuelva la similaridad entre ambos vectores
###
### Paso 1: Si u y v son listas -> Convertirlo a arrays
###
### Paso 2: Calcula la similaridad entre ambos vectores
### TU RESPUESTA ABAJO

import numpy as np

def cosine_similarity(u, v):
    """Calcula la similaridad entre dos vectores
    
    Argumentos:
        u -- Vector 1
        v -- Vector 2
        
    Ejemplo:
        u = np.array([1, 2, 3])
        v = np.array([3, 2, 1])
        
        similarity = cosine_similarity(u, v)  # -> 0.71428
    """
    # Si u y v son listas, convertirlo a arrays
    u = np.array(u)
    v = np.array(v)
    
    # Calcula la similaridad entre ambos vectores
    similarity = np.dot(u, v) / (np.linalg.norm(u) * np.linalg.norm(v))
    
    return similarity

u = np.array([1, 2, 3])
v = np.array([3, 2, 1])
cosine_similarity(u, v)
```
